```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
values = ['Alpha', 'Beta', 'Gamma', 'Delta', 'Epsilon', 'Zeta']
X = np.random.choice(values, size=(100, 1))
y = np.random.randint(6, size=(100, 1))
data = np.hstack((X, y))

# Create dataframe
df = pd.DataFrame(data, columns=['Category', 'Value'])
```

```python
df.head()
```

```python
# Plot

# Colors (for each of 6 columns)
rgb_values = sns.color_palette('Set1', 6)
color_labels = df['Category'].unique()
color_map = dict(zip(color_labels, rgb_values))


# Plot
fig, ax = plt.subplots(figsize=(16, 10))

for i in color_map.items():
    plt.scatter(x=df['Category'].index.values, y=df['Value'], c=df['Category'].map(color_map),
               s=150, edgecolors='k', label=i[0])
    

# Legend
plt.legend(prop={'size' : 20})
ax = plt.gca()
legend = ax.get_legend()

for i, j in enumerate(rgb_values):
    legend.legendHandles[i].set_color(j)
    

# Show grid
plt.grid(True)

# Title
plt.title('Randomly Generated Scatter - With Color Coded Label', fontsize=24)

    
plt.show()
```

```python
# Subplots with color coded legends


# Colors (for each of 6 columns)
rgb_values = sns.color_palette('Set2', 6)
color_labels = df['Category'].unique()
color_map = dict(zip(color_labels, rgb_values))


# Plots (unpack tuples instead of ax[0,0], ax[0,1], etc.)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(26, 20))

for i in color_map.items():
    ax1.scatter(x=df['Category'].index.values, y=df['Value'], c=df['Category'].map(color_map),
                s=150, edgecolors='k', label=i[0])
    ax1.legend()
    ax2.scatter(x=df.loc[df['Category'] == 'Alpha'].index.values, y=df['Value'].loc[df['Category'] == 'Alpha'], 
                c=df['Category'].loc[df['Category'] == 'Alpha'].map(color_map),
                s=150, edgecolors='k', label=i[0])
    ax2.legend()
    ax3.scatter(x=df.loc[df['Category'] == 'Beta'].index.values, y=df['Value'].loc[df['Category'] == 'Beta'], 
                c=df['Category'].loc[df['Category'] == 'Beta'].map(color_map),
                s=150, edgecolors='k', label=i[0])
    ax3.legend()
    ax4.scatter(x=df.loc[df['Category'] == 'Gamma'].index.values, y=df['Value'].loc[df['Category'] == 'Gamma'], 
                c=df['Category'].loc[df['Category'] == 'Gamma'].map(color_map),
                s=150, edgecolors='k', label=i[0])
    ax4.legend()
    

# Titles
ax1.title.set_text('Randomly Generated Subplots - Color LabeledLegends')
ax1.title.set_fontsize(18)

    
# Get all axes, iterate thru legendHandles and change colors
for k in plt.gcf().get_axes():
    plt.sca(k) #Set current axis
    ax = plt.gca()
    legend = ax.get_legend()
    k.tick_params(labelsize=20) #Set tick sizes on axes
    
    for i, j in enumerate(rgb_values):
        legend.legendHandles[i].set_color(j)
        
        
        
# Change color for specific legend label
# plt.sca(ax2)
# plt.legend()
# ax = plt.gca()
# leg = ax.get_legend()
# leg.legendHandles[0].set_color('green')
    
    
plt.show()
```
